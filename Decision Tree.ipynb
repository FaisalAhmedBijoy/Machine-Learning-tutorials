```python
import pandas as pd
df=pd.read_csv('decisionsalary.csv')
df
```

```python
inputs=df.drop(['salary100k'],axis='columns')
inputs
```

```python
target=df.salary100k
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder

le_company=LabelEncoder()
le_job=LabelEncoder() 
le_degree=LabelEncoder()
```

```python
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_job.fit_transform(inputs['job'])
inputs['degree_n']=le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,1,0]])
```

```python
df=pd.read_csv('titanic_data.csv')
df.head()
```

```python
df.Age.mean()
```

```python
df.Age=df.Age.fillna(df.Age.mean())
```

```python
df.Age
```

```python
df.head()
```

```python
inputs=df.drop(['passengerid','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
inputs
```

```python
target=df.Survived
```

```python
target
```

```python
inputs
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
inputs
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target)
```

```python
x_train
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[1,1,25,100]])
```
