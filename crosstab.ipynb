```python
import pandas as pd
df=pd.read_excel('crosstab.xlsx')
df
```

```python
pd.crosstab(df.nationality,df.handedness)
```

```python
pd.crosstab(df.sex,df.handedness)
```

```python
pd.crosstab(df.sex,[df.handedness,df.nationality])
```

```python
pd.crosstab(df.handedness,df.age)
```

```python
pd.crosstab(df.handedness,df.sex)
```

```python

pd.crosstab([df.nationality,df.sex],df.handedness,margins=True)
```

```python
pd.crosstab([df.sex],[df.handedness],margins=True,normalize='index')
```

```python
import numpy as np
pd.crosstab([df.sex],[df.handedness],values=df.age,aggfunc=np.average)
```
