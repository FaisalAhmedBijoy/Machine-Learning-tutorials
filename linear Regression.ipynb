```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df=pd.read_csv('homerent.csv')
df
```

```python
plt.plot(df)
```

```python
plt.xlabel('area')
plt.ylabel('US $ ')
plt.title('Home Rent')
plt.scatter(df.area,df.price,color='red')

```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
plt.scatter(df.area,df.price,color='red')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
reg.predict([[3300]])
```

```python
area_df=pd.read_csv('homeareas.csv')
area_df
```

```python
p=reg.predict(area_df)
p
```

```python
area_df['price']=p
area_df
```

```python
area_df.to_csv('predictionhome.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

df=pd.read_csv('multiplereg.csv')
df
```

```python
df.bedroom.mean()
```

```python
import math
median_bedroom=math.floor(df.bedroom.mean())
median_bedroom
df.bedroom=df.bedroom.fillna(median_bedroom)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df.price)
plt.xlabel('Area')
plt.ylabel('US $')
plt.title('Home Rent in USA ')
plt.scatter(df.area,df.bedroom,df.age,df.price)
plt.plot(df.area,df.price,label='price')
plt.plot(df.area,df.bedroom,label='bedroom')
plt.plot(df.area,df.age,label='age')
plt.legend()
reg.predict([[3000,3,20]])
plt.plot(df.area,df.bedroom,df.age,reg.predict(df[['area','bedroom','age']]))
```
